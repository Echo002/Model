```python
import  matplotlib.pyplot as plt
import random
import matplotlib
```

```python
#生成数据
def data():
    x = range(0,50)
    y = [(3*s+6) for s in x]
    for i in range(0,50):
        y[i] = y[i]+random.randint(0,30)
    return x, y
```

```python
x,y = data()
matplotlib.rcParams['font.sans-serif'] = ['SimHei'] #设置字体中文，防止乱码
plt.scatter(x, y, color = 'red',label = '数据')
plt.legend()
plt.show()
```

```python
def SGD_function(x,y):
    error0 = 0
    step_size = 0.001
    esp = 1e-6
    a = random.randint(0,4)
    b = random.randint(0,8)
    m = len(x)
    n = 0
    while True:
        i = random.randint(0,m-1)
        print(i)
        sum0 = a * x[i] + b - y[i]
        sum1 = (a * x[i] + b - y[i])*x[i]
        error1 = (a * x[i] + b - y[i])**2
 
        a = a - sum1*step_size/m
        b = b - sum0*step_size/m
        print('a=%f,b=%f,error=%f'%(a,b,error1))
 
        if abs(error1-error0)<esp:  #误差很小，可以终止迭代
            break
        error0 = error1
        n = n+1
        if n%50==0:
            print('第%d次迭代'%n)
        if (n>1000):
            break
    return a,b
```

```python
x,y = data()
a,b = SGD(x,y)
X = range(50)
Y = [(a*i+b) for i in X]
 
plt.scatter(x,y,color='red')
plt.plot(X,Y)
plt.show()
```
